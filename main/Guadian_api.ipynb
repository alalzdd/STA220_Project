```python
import requests
import json
from datetime import date, datetime
import os
import time

API_KEY = os.environ.get("API_KEY")
query = "ukraine war"
from_date = "2022-01-01"  # Start date
to_date = date.today().strftime("%Y-%m-%d")  # End date (today)

# The maximum allowed page size by The Guardian API is 200
page_size = 200

# Dictionary to hold the results, with article's title as the key
# and a list [publication_date, url] as the value
all_results = {}

page = 1  # We start from page 1

while True:
    # Build the URL for the current page
    url = (
        "https://content.guardianapis.com/search"
        f"?q={query}"
        f"&from-date={from_date}"
        f"&to-date={to_date}"
        f"&order-by=newest"
        f"&page={page}"
        f"&page-size={page_size}"  # <-- Note the hyphen here
        f"&api-key={API_KEY}"
    )

    response = requests.get(url)
    # wait for one second to avoid ratelimit
    time.sleep(1)
    if response.status_code != 200:
        print(f"Error: received status code {response.status_code}")
        break

    data = response.json()

    # Safety check in case the JSON doesn't have the expected structure
    if "response" not in data or "results" not in data["response"]:
        print("No valid 'response' or 'results' field in JSON.")
        break

    current_results = data["response"]["results"]

    # Add them to our dictionary
    for article in current_results:
        title = article["webTitle"]
        pub_date_str = article["webPublicationDate"]  # e.g. "2023-02-10T12:34:56Z"

        # Convert to a datetime object
        # (ISO 8601 with trailing 'Z' means UTC, so we replace 'Z' with '+00:00')
        pub_date_dt = datetime.fromisoformat(pub_date_str.replace("Z", "+00:00"))
        
        # If you want to store just 'YYYY-MM-DD' as a string, do:
        pub_date_formatted = pub_date_dt.strftime("%Y-%m-%d")

        web_url = article["webUrl"]
        
        # Store in the dictionary: title -> [date, web_url]
        # NOTE: If multiple articles have the same title, they will overwrite each other
        all_results[title] = [pub_date_formatted, web_url]

    current_page = data["response"].get("currentPage", 1)
    total_pages = data["response"].get("pages", 1)

    print(
        f"Fetched page {current_page}/{total_pages}. " 
        f"Articles so far: {len(all_results)}",
        end="\r"
    )

    # Stop if we've reached the last page
    if current_page >= total_pages:
        break

    # Otherwise, go to the next page
    page += 1

# Once all pages are fetched, save results as a JSON file
with open("../data/guardian_ukraine_war_headlines.json", "w", encoding="utf-8") as f:
    json.dump(all_results, f, ensure_ascii=False, indent=2)

print(f"\nDone! Saved {len(all_results)} articles to guardian_ukraine_war_headlines.json")
```
